```python
# Refer to the numericals.csv dataset that you previously saved.
# Use the numerical variables from our dataset this week.
import pandas as pd
import numpy as np

data = pd.read_csv('numerical2.csv') 
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
# Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,20))
for i, column in enumerate(data.columns, 1):
    plt.subplot(5,4,i)
    sns.histplot(data[column])
```

```python
def log_transformer(x):
    if np.isfinite(x) and x!=0: 
        return np.log(x)
    else:
        return np.NAN
```

```python
for i in data.columns:
    data[i]=list(map(log_transformer,data[i]))
```

```python
sns.set(rc={'figure.figsize':(5,4)})
for i in data.columns:
    sns.distplot(data[i])
    plt.show()
```
